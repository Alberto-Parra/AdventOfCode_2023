```python
import pandas as pd
```

```python
df=pd.DataFrame()
df=pd.read_csv('puzzle_input.csv',sep=';',header=None)
df
```

```python

possibles=[]
for game in range(df.shape[0]):
    possible=True
    blue=0
    green=0
    red=0
    for samples in range(1,6,1):
        if type(df[samples][game])==type('str'):
            sample=df[samples][game]
            sample=sample.split()
            
            for i in range(0,len(sample),2):
                if sample[i+1]=='green' or sample[i+1]=='green,':
                    green=int(sample[i])
                if sample[i+1]=='blue' or sample[i+1]=='blue,':
                    blue=int(sample[i])
                if sample[i+1]=='red' or sample[i+1]=='red,':
                    red=int(sample[i])

            if blue>14 or red>12 or green>13:
                possible=False
                break
            else:
                possible=True
        else:
            pass
        
    if possible:    
        possibles.append(game+1)
        

sum(possibles)
```
