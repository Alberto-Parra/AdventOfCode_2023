```python
import pandas as pd
```

```python
df=pd.DataFrame()
df=pd.read_csv('puzzle_input.csv',sep=';',header=None)
df
```

```python
possibles=[]
for game in range(df.shape[0]):
    red=[]
    blue=[]
    green=[]
    for samples in range(1,6,1):
        if type(df[samples][game])==type('str'):
            sample=df[samples][game]
            sample=sample.split()
            print(sample)
            for i in range(len(sample)-1):
                if sample[i+1]=='green':
                    green.append(int(sample[i]))
                elif sample[i+1]=='blue':
                    blue.append(int(sample[i]))
                elif sample[i+1]=='red':
                    red.append(int(sample[i]))
        else:
            pass
        
    if sum(blue)<=14:
        if sum(red)<=12:
            if sum(green)<=13:
                possibles.append(game)

possibles
```
